### defining attribute and method

```python
class Human():

    def __init__(self):
        self.h = 175
        self.w = 70

    def running(self):
        pass
```

### definning and calling a method

```python
class Human3():

    def __init__(self, first, last, height, weight):
        self.first = first
        self.last = last
        self.h = height
        self.w = weight

    def introduce(self):
        print(f"My name is {self.first} {self.last}. I am {self.h} cm in height and {self.w} kg in weight")
```

## Instance variable vs Class variable

```python
class Human():
    
    raise_amount = 1.05

    def __init__(self, first, last, height, weight):
        self.first = first
        self.last = last
        self.h = height
        self.w = weight

    def introduce(self):
        print(f"My name is {self.first} {self.last}.")

    def overweight(self):
        print(self.w * self.raise_amount)
        
```

```python
human_1 = Human("M", "B", 180, 70)
human_1.introduce()
human_1.overweight()
```

```python
print(Human.raise_amount)
print(human_1.raise_amount)
```

#### the following example shows the importance of how we call the class variable in a method. In the example we are investigating, There is a subtke difference between     ((self.w = self.w * Human.raise_amount))     and    ((self.w = self.w * self.raise_amount)). in the first one, it always referes to the Human class to read the value of raise_amount

```python
human_1.raise_amount = -2
print(Human.raise_amount)
print(human_1.raise_amount)
human_1.overweight()
```

```python
human_2 = Human("a", "A", 123, 45)
human_3 = Human("b", "B", 99, 9)

print(human_2.raise_amount)
human_2.raise_amount = -2
print(human_2.raise_amount)
print(human_3.raise_amount)
print(Human.raise_amount)
```

### Another example to better figure out the difference between self. and Human.

```python
class Human():

    raise_amount = 1.05
    number_of_instance = 0

    def __init__(self, first, last, height, weight):
        self.first = first
        self.last = last
        self.h = height
        self.w = weight
        Human.number_of_instance += 1

    def introduce(self):
        print(f"My name is {self.first} {self.last}.")

    def overweight(self):
        print(self.w * self.raise_amount)
```

```python
human_2 = Human("a", "A", 123, 45)
print(Human.number_of_instance)
human_3 = Human("b", "B", 99, 9)
print(Human.number_of_instance)
```

The regular (instance) method is like the "introduce" method that used in the Human class. Now let's see the class method.

# Class Method

```python
class Human():

    raise_amount = 1.05
    number_of_instance = 0

    def __init__(self, first, last, height, weight):
        self.first = first
        self.last = last
        self.h = height
        self.w = weight
        Human.number_of_instance += 1

    def introduce(self):
        print(f"My name is {self.first} {self.last}.")

    def overweight(self):
        print(self.w * self.raise_amount)

    @classmethod
    def set_raise_amount(cls, newValue):
        cls.raise_amount = newValue
```

```python
human_1 = Human("A", "a", 100, 50)

human_1.introduce()
print(Human.number_of_instance)
print(human_1.raise_amount)
print(Human.raise_amount)

Human.set_raise_amount(-2)
print(human_1.raise_amount)
print(Human.raise_amount)
```
