## Instance variable vs Class variable

```python
class Human():
    
    raise_amount = 1.05

    def __init__(self, first, last, height, weight):
        self.first = first
        self.last = last
        self.h = height
        self.w = weight

    def introduce(self):
        print(f"My name is {self.first} {self.last}.")

    def overweight(self):
        print(self.w * self.raise_amount)
        
```

```python
human_1 = Human("M", "B", 180, 70)
human_1.introduce()
human_1.overweight()
```

```python
print(Human.raise_amount)
print(human_1.raise_amount)
```

In the example we are investigating, There is a subtke difference between     ((self.w = self.w * Human.raise_amount))     and    ((self.w = self.w * self.raise_amount)). in the first one, it always referes to the Human class to read the value of raise_amount

```python
human_1.raise_amount = -2
print(Human.raise_amount)
print(human_1.raise_amount)
human_1.overweight()
```

# Class Method

```python
class Human():

    raise_amount = 1.05
    number_of_instance = 0

    def __init__(self, first, last, height, weight):
        self.first = first
        self.last = last
        self.h = height
        self.w = weight
        Human.number_of_instance += 1

    def introduce(self):
        print(f"My name is {self.first} {self.last}.")

    def overweight(self):
        print(self.w * self.raise_amount)

    @classmethod
    def set_raise_amount(cls, newValue):
        cls.raise_amount = newValue
```

```python
human_1 = Human("A", "a", 100, 50)

human_1.introduce()
print(Human.number_of_instance)
print(human_1.raise_amount)
print(Human.raise_amount)

Human.set_raise_amount(-2)
print(human_1.raise_amount)
print(Human.raise_amount)
```

## Class method as an alternative constructor

```python
class Human():

    raise_amount = 1.05
    number_of_instance = 0

    def __init__(self, first, last, height, weight):
        self.first = first
        self.last = last
        self.h = height
        self.w = weight
        Human.number_of_instance += 1

    def introduce(self):
        print(f"My name is {self.first} {self.last}.")

    def overweight(self):
        print(self.w * self.raise_amount)

    @classmethod
    def set_raise_amount(cls, newValue):
        cls.raise_amount = newValue

    @classmethod
    def from_string(cls, input_str):
        first, last, h, w = input_str.split("-")
        h, w = int(h), int(w)
        return cls(first, last, h, w)
    
human_1 = "Mammad-Bhr-175-70"
human_1 = Human.from_string(human_1)
human_1.introduce()
```

## Static Method

```python
from datetime import date

class Human():

    raise_amount = 1.05
    number_of_instance = 0

    def __init__(self, first, last, height, weight):
        self.first = first
        self.last = last
        self.h = height
        self.w = weight
        Human.number_of_instance += 1

    def introduce(self):
        print(f"My name is {self.first} {self.last}.")

    def overweight(self):
        print(self.w * self.raise_amount)

    @classmethod
    def set_raise_amount(cls, newValue):
        cls.raise_amount = newValue

    @classmethod
    def from_string(cls, input_str):
        first, last, h, w = input_str.split("-")
        h, w = int(h), int(w)
        return cls(first, last, h, w)
    
    @staticmethod
    def creation_date():
        print(date.today())

human_1 = "Mammad-Bhr-175-70"
human_1 = Human.from_string(human_1)
human_1.introduce()
human_1.creation_date()
```

## Example
Create a class called Product that tracks inventory and applies discounts.

Class Attribute: Define a variable tax_rate set to 0.05 (5%).

Instance Method: Create get_total_price(self, quantity). It should calculate the price for a specific number of items, including the tax_rate.

Class Method: Create set_tax_rate(cls, new_rate). This should allow you to update the tax rate for all products at once.

Static Method: Create is_valid_price(price). This should simply return True if a price is greater than zero, and False otherwise (it doesn't need to know about the class or the instance).

```python
class Product():

    tax_rate = 0.05

    def __init__(self, name, price):
        self.name = name
        self.price = price

    def total_price(self, quantity):
        self.tot_price = (1+Product.tax_rate) * (self.price*quantity)
        return self.tot_price


    @classmethod
    def change_tax_rate(cls, new_value):
        cls.tax_rate = new_value

    @staticmethod
    def is_valid(price):
        return price > 0


p1 = Product("A", 100)
print(p1.total_price(2) )

print(Product.tax_rate)
Product.change_tax_rate(0.1)
print(Product.tax_rate)
print(p1.tax_rate)
print(p1.total_price(2))
p1.is_valid(p1.total_price(2))
```

## The Example: BankAccount Class
Instance Method (deposit): Operates on a specific user's balance. It needs self to know which account to update.

Class Method (set_interest_rate): Operates on the class variable. If the bank changes its global interest rate, every account is affected.

Static Method (is_valid_account_number): A utility check. It validates if a string follows a specific format (e.g., 10 digits) without needing to access any specific account or bank data.

```python
class BnakAccount():

    interest_rate = 0.02

    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"added {amount} and new balance now is {self.balance}")

    @classmethod
    def set_interest_rate(cls, new_rate):
        cls.interest_rate = new_rate

    @staticmethod
    def is_valid(num):
        print(len(str(num)) == 10 and str(num).isdigit() == True)


client_1 = BnakAccount("A", 100)
print(client_1.balance)
print(client_1.interest_rate)

client_1.deposit(25)

client_1.set_interest_rate(0.025)
print(BnakAccount.interest_rate)
print(client_1.interest_rate)

client_1.is_valid("0123456789")
client_1.is_valid("012345678s")
```

### If we want new attributes in addition to the parent attributes:

```python
class Human():
    
    raise_amount = 1.05

    def __init__(self, first, last, height, weight):
        self.first = first
        self.last = last
        self.h = height
        self.w = weight

    def introduce(self):
        print(f"My name is {self.first} {self.last}.")

    def overweight(self):
        print(self.w * self.raise_amount)

    @classmethod
    def from_str(cls, str_input):
        first, last, h, w = str_input.split("-")
        h, w = int(h), int(w)
        return cls(first, last, h, w)

class Man(Human):
    raise_amount = 1.2

    def __init__(self, first, last, height, weight, marriage):
        super().__init__(first, last, height, weight)
        self.marriage = marriage

    @classmethod
    def from_str(cls, str_input):
        first, last, h, w, marriage = str_input.split("-")
        h, w = int(h), int(w)
        return cls(first, last, h, w, marriage)


man_1 = Man.from_str("A-a-100-50-30")
human_1 = Human.from_str("B-b-70-30")
man_2 = Man("Z", "z", 123, 23, 33)

man_1.introduce()
human_1.introduce()

print(man_1.raise_amount)
print(human_1.raise_amount)
print(man_1.marriage)

man_2.introduce()
print(man_2.marriage)
```
