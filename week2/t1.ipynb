## Item 9: Avoid else after for and while

```python
for i in range(3):
    print(i)
else:
    print("else block")
```

```python
for i in range(5):
    print(i)
    if i == 3:
        break
else: 
    print("else block")
```

```python
for i in []:
    print("never runs")
else:
    print("else block")
```

```python
while False:
    print("never runs")
else:
    print("else block")
```

```python
a = 7
b = 21

for i in range(2, min(a, b) + 1):
    print(f"testing {i}")
    if a % i == 0 and b % i == 0:
        print("\nnot coprime")
        break
else:
    print("\ncoprime")
```

```python
a = 8
b = 21

def coprime(a, b):
    for i in range(2, min(a, b) + 1):
        if a % i == 0 and b % i == 0:
            return False
    return True

coprime(a, b)
```

```python
a = 8
b = 21

is_prime = True

for i in range(2, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        is_prime = False
        break
print(is_prime)
```

## Item 10: Prevent repetition with assignment expressions


```python
fresh_fruit = {
    'apple' : 3,
    'banana' : 5,
    'lemon' : 4
}
```

### using assignment operator (=)

```python
def make_lemonade():
    print("making lemonade in progress . . .")

def out_of_stock():
    print("ran out of lemon")

count = fresh_fruit.get('lemon', 0)
if count:
    make_lemonade()
else:
    out_of_stock()
```

### using assignment expression (:=) ------> walrus

```python
def make_lemonade(count):
    print(f"there are {count} lemons")

def out_of_stock():
    print("ran out of lemon")

if count := fresh_fruit.get('lemon', 0):
    make_lemonade(count)
else:
    out_of_stock()
```

```python
def make_cider(count):
    print(f"there are {count} apples")

def out_of_stock():
    print("not enough apple")

if (count := fresh_fruit.get('apple', 0)) >= 4:
    make_cider(count)
else:
    out_of_stock()
```
