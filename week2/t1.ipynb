```python
a = 7
b = 21

for i in range(2, min(a, b) + 1):
    print(f"testing {i}")
    if a % i == 0 and b % i == 0:
        print("\nnot coprime")
        break
else:
    print("\ncoprime")
```

```python
a = 8
b = 21

def coprime(a, b):
    for i in range(2, min(a, b) + 1):
        if a % i == 0 and b % i == 0:
            return False
    return True

coprime(a, b)
```

```python
a = 8
b = 21

is_prime = True

for i in range(2, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        is_prime = False
        break
print(is_prime)
```

## Item 10: Prevent repetition with assignment expressions


```python
fresh_fruit = {
    'apple' : 3,
    'banana' : 5,
    'lemon' : 4
}
```

### using assignment operator (=)

```python
def make_lemonade():
    print("making lemonade in progress . . .")

def out_of_stock():
    print("ran out of lemon")

count = fresh_fruit.get('lemon', 0)
if count:
    make_lemonade()
else:
    out_of_stock()
```

### using assignment expression (:=) ------> walrus

```python
def make_lemonade(count):
    print(f"there are {count} lemons")

def out_of_stock():
    print("ran out of lemon")

if count := fresh_fruit.get('lemon', 0):
    make_lemonade(count)
else:
    out_of_stock()
```

```python
def make_cider(count):
    print(f"there are {count} apples")

def out_of_stock():
    print("not enough apple")

if (count := fresh_fruit.get('apple', 0)) >= 4:
    make_cider(count)
else:
    out_of_stock()
```

#### another example
first without walrus

```python
fresh_fruit = {
    'apple' : 3,
    'banana' : 5,
    'lemon' : 4
}

def smoothie(count):
    print(f"there are {count} bananas to make smoothie")

def lemonade(count):
    print(f"there are {count} lemons to make lemonade")

def cider(count):
    print(f"there are {count} apples to make cider")

count = fresh_fruit.get('banana', 0)
if count >= 2:
    smoothie(count)
else:
    count = fresh_fruit.get('apple', 0)
    if count >= 4:
        cider(count)
    else:
        count = fresh_fruit.get('lemon', 0)
        if count >= 2:
            lemonade(count)
        else:
            print("out of stock")
```

```python
fresh_fruit = {
    'apple' : 1,
    'banana' : 1,
    'lemon' : 1
}

def smoothie(count):
    print(f"there are {count} bananas to make smoothie")

def lemonade(count):
    print(f"there are {count} lemons to make lemonade")

def cider(count):
    print(f"there are {count} apples to make cider")

if (count := fresh_fruit.get('banana', 0)) >= 2:
    smoothie(count)
elif (count := fresh_fruit.get('apple', 0)) >= 4:
    cider(count)
elif (count := fresh_fruit.get('lemon', 0)) >= 2:
    lemonade(count)
else:
    print("out of stock")
```

## Item 11: Know how to slice sequences

```python
a = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']

print("Middle two: ", a[3 : 5])
print("All but ends: ", a[1: 7])
```

### Do NOT use 0 for start and len for end; use : instead.

```python
# instead of
print(a[0:5])
# say
print(a[:5])

```

### The list resulting from slicing is independent of the main list

```python
a = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
b = a[3:]

print("b beofre:  ", b)
b[1] = 100
print("b after:   ", b)
print("a:   ", a)
```

### Using slicing in value assignment, the number of values in two sides can differ from each other

```python
a = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
print("a before:    ", a)
a[2:6] = [100]
print("a after:     ", a)
```
